```python
import os
from bs4 import BeautifulSoup as bs
from splinter import Browser
import requests
import pandas as pd
```

```python
household_size_url = 'https://www.indexmundi.com/facts/united-states/quick-facts/all-states/average-household-size#table'
```

```python
household_size = pd.read_html(household_size_url)
household_size_df = household_size[0]
household_size_df.columns = ['State', 'Household Size']
household_size_df.head()
```

```python
median_income_url = 'https://www.indexmundi.com/facts/united-states/quick-facts/all-states/median-household-income#table'
```

```python
median_income = pd.read_html(median_income_url)
median_income_df = median_income[0]
median_income_df.columns = ['State', 'Median Income']
median_income_df.head()
```

```python
poverty_url = 'https://www.indexmundi.com/facts/united-states/quick-facts/all-states/percent-of-people-of-all-ages-in-poverty#table'
poverty = pd.read_html(poverty_url)
poverty_df = poverty[0]
poverty_df.columns = ['State', 'Poverty']
poverty_df.head()
```

```python
combined_df = pd.merge(household_size_df, median_income_df, how='left', on ='State' )
combined_df
```

```python
combined_df.set_index('State', inplace =True)
combined_df.head()
```

```python
household = combined_df.to_csv('household.csv', index = True)
```
